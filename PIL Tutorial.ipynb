PIL Tutorial

```python
from PIL import Image
import os #to loop over files in directory
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1=Image.open('Arcane/1.jpg')
img1.show()
img2=Image.open('Arcane/3.jpg')
img2.show()
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
print(img1.size)
print(img1.format)
print(img1.mode)
```

```python
img1.save('Arcane/Saving_1.jpg')  #saving image as other name in directory
```

```python
left=500
top=300
bottom=500
right=700
crp_img=img2.crop((left,top,right,bottom))
plt.imshow(crp_img)
crp_img.save('Arcane/Cropped_2.jpg')
```

```python
img3=crp_img.copy()
plt.imshow(img3)
```

```python
newsize=(300,300)

rszimg1=img3.resize(newsize,Image.BICUBIC)
plt.subplot(1,3,1)
plt.imshow(rszimg1)

rszimg2=img3.resize(newsize,Image.NEAREST)
plt.subplot(1,3,2)
plt.imshow(rszimg2)

rszimg3=img3.resize(newsize,Image.HAMMING)
plt.subplot(1,3,3)
plt.imshow(rszimg3)
```

```python
angle=30
rot_img1=img1.rotate(angle)
plt.imshow(rot_img1)
```

```python
bw_img1=img1.convert('L')  # pass string L to convert it into black and white
plt.subplot(2,1,1)
plt.imshow(bw_img1)
plt.subplot(2,1,2)
plt.imshow(bw_img1, cmap='gray')  # use cmap=gray for matplotlib to correctly show black and white
```

```python
hsv_img1=img1.convert('HSV')
plt.imshow(hsv_img1)
```

```python
array=np.array(img1)
print(array.shape)
print(array)
#converts image to numpy array
```

```python
image=Image.fromarray(array)#converts numpy array to image
plt.imshow(image)
```

```python
from PIL import ImageEnhance
col_img1=ImageEnhance.Brightness(img1).enhance(2.5)
plt.subplot(1,2,1)
plt.imshow(col_img1)

col2_img1=ImageEnhance.Sharpness(img1).enhance(4.5)
plt.subplot(1,2,2)
plt.imshow(col2_img1)
col2_img1.save('Arcane/1_sharp.jpg')
```

```python
# out=img1*(1-0.3)+img2*0.3  images should be of same dimensions

blend_img=Image.blend(img1,img2,0.5)
plt.imshow(blend_img)
```

```python
img_channel=img1.copy()
r,g,b=img_channel.split()
plt.subplot(4,1,1)
plt.imshow(r)
plt.subplot(4,1,2)
plt.imshow(g)
plt.subplot(4,1,3)
plt.imshow(b)
plt.subplot(4,1,4)
plt.imshow(Image.merge("RGB",(r,g,b)))
```
